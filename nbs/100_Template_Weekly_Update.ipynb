# Weekly 5-Minute Project Update Template

**Goal:** In 5 minutes, communicate progress, evidence, blockers, and the plan for next week.  
**Format:** 4–6 slides max (or 1 slide + spoken structure). Aim for clarity over completeness.

---

### Slide 1 Project Snapshot
- **Project title**
- Group members
- One-sentence goal: *We are building X to achieve Y in scenario Z.*
- Current status: on track / at risk / blocked
- Milestone target (this week):** *What you intended to achieve*

---

### Slide 2 What We Did Since Last Week
3–5 bullets, phrased as outcomes (not just activities):
- Implemented / tested / validated …
- Integrated / tuned / debugged …
- Produced: *plot / table / dataset / simulation results / code module*

*(If helpful, include 1 small screenshot or figure.)*

---

### Slide 3 Evidence: One Key Result
Show **one** figure/table only:
- What does the plot show?
- What is the main takeaway in **one sentence**?
- Quantify if possible: RMSE, max error, convergence rate, stability, runtime, etc.

**Figure caption template:**  
*Scenario: ___ | Metric: ___ | Baseline: ___ | Result: ___*

---

### Slide 4 Issues / Risks / Questions 
- **Blockers (if any):** what is stopping progress?
- **Risks:** what might break later (model assumptions, parameter uncertainty, noise sensitivity)?
- **Help needed:** a specific question to the instructor or class

*(Keep it concrete: “We need guidance on selecting damping terms” > “We are unsure.”)*

---

### Slide 5 Plan for Next Week 
- **Top 3 tasks** 
  1. …
  2. …
  3. …
- **Expected output:** *plot / experiment / module / report section / demo*
- **Success criteria:** *how you’ll know it worked*

---

### Slide 6 (Optional) Timeline / Task Board 
- Next milestone date
- Who is doing what (1 line each)
- Any dependencies

---

## Spoken Script (No Slides Version, 5-Minute Structure)

1. **Snapshot:**  
   “Our project is ___. This week we aimed to ___. Status is __.”

2. **Progress:**  
   “Since last week, we completed: (1) ___, (2) ___, (3) ___.”

3. **Evidence:**  
   “The key result is ___. This figure shows ___. The main takeaway is ___. Quantitatively, ___.”

4. **Problems / help needed:**  
   “We are currently blocked/risked by ___. We’d like feedback on ___.”

5. **Next steps:**  
   “Next week we will deliver ___. Success means ___.”

---

## Weekly Update Checklist 
Before Presenting make sure you have:

- [ ] One clear takeaway for the week
- [ ] One plot/table (max) with labeled axes and units
- [ ] A baseline or reference (even if simple)
- [ ] One concrete question for feedback (if needed)
- [ ] Next-week plan has deliverables + success criteria
